## Basics zu Jupyter Notebooks
Jupyter Notebooks bestehen aus 2 Arten von Blöcken (sogenannte Zellen): Code- und Textblöcke. Sie können beliebig viele Zellen hinzufügen und wieder löschen.

Textblöcke befinden sich immer in einem von 2 Modi: 
- Editier-Modus (sieht z.T. kryptisch aus): Durch Doppelklick auf eine Text-Zelle wechselt die Zelle in den Editiermodus.
- Anzeige-Modus (sieht etwas schöner aus): Durch Drücken von ctrl+Enter wechselt die Zelle in den Anzeigemodus.

Innerhalb von Textblöcken können sie einfache Formatierungs-Anweisungen (Markdown-Syntax) verwenden, um den Text zu gestalten. 
- Um ein Wort **fett** zu schreiben können Sie das Wort im Editier-Modus doppelklicken, dann 2x Shift+3 (ergibt 2 Sternchen * vor und hinter dem Wort). Dadurch wird ein Wort oder Textabschnitt fett dargestellt. 
- Um etwas kursiv zu schreiben *unterstreichen*
Weitere Inputs dazu finden Sie im Web, z.B. [hier](https://www.markdownguide.org/cheat-sheet/).

Möchten Sie Python Bibliotheken (sog. Packages) importieren, könnne Sie das wie in Python Skripten gewohnt über die import Statements tun.

```python
import numpy as np
import matplotlib.pyplot as plt

# Wertebereich definieren
x = np.linspace(0, 2 * np.pi, 1000)  # Von 0 bis 2π, 1000 Punkte
y = np.sin(x)  # Sinusfunktion

# Plot erstellen
plt.plot(x, y, label='sin(x)', color='blue')
plt.title('Plot der Sinusfunktion')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## Apps in Jupyter Notebooks
In Jupyter Notebooks können Sie auch einfach kleine Apps programmieren. Die Möglichkeiten sind sehr limitiert und nicht vergleichbar mit richtigen UI-Frameworks, alles basiert auf ipywidgets und die GUIs werden deklarativ zusammengestellt. Dennoch reichen die Möglichkeiten für einfache Aufgabenstellungen aus.

Nachfolgend ein simples Beispiel:

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from verdoppeln import verdopple
# Widgets definieren
hinweis_label = widgets.Label(value="Bitte eine Zahl wählen:")

# Slider für die Eingabe einer ganzen Zahl
my_integer_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=10,
    step=1,
    description="Zahl n:",
    style={"description_width": "initial"},
    layout=widgets.Layout(width="400px")
)

# Output-Widget für die Anzeige des Ergebnisses
output = widgets.Output()

# Funktion, die aufgerufen wird, wenn der Button geklickt wird
def button_handler(c):
    result = verdopple(my_integer_slider.value)
    output.clear_output()
    with output:
        print(f"Das Ergebnis ist: {result}")

button = widgets.Button(description="Verdopple n", button_style="success")
button.on_click(button_handler)

# Anzeige der Widgets
display(widgets.VBox([hinweis_label, my_integer_slider, button, output]))
```
